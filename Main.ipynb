```python
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

# Create a copy for preprocessing
df_processed = df.copy()
```

```python
# Standardize 'Amount' and 'Time'
scaler = StandardScaler()
df_processed[['scaled_amount', 'scaled_time']] = scaler.fit_transform(df_processed[['Amount', 'Time']])
```

```python
# Drop original 'Amount' and 'Time'
df_processed.drop(['Amount', 'Time'], axis=1, inplace=True)
```

```python
# Move the scaled columns to the front for clarity
scaled_features = ['scaled_amount', 'scaled_time']
remaining_features = [col for col in df_processed.columns if col not in scaled_features + ['Class']]
ordered_columns = scaled_features + remaining_features + ['Class']
df_processed = df_processed[ordered_columns]
```

```python
# Split features and target
X = df_processed.drop('Class', axis=1)
y = df_processed['Class']
```

```python
# Apply SMOTE to balance the dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Check new class distribution
resampled_class_distribution = y_resampled.value_counts()
resampled_class_distribution
```

```python
# Train-Test Split on Resampled Data
from sklearn.model_selection import train_test_split

# Split resampled data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)
```

```python
# Train Classification Models
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)
y_pred_log = log_model.predict(X_test)
y_proba_log = log_model.predict_proba(X_test)[:, 1]


from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
y_proba_rf = rf_model.predict_proba(X_test)[:, 1]

```

```python
# Evaluate models
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report

def evaluate_model(name, y_test, y_pred, y_proba):
    print(f"\nüìä Evaluation Metrics for {name}:")
    print("Accuracy :", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))
    print("Recall   :", recall_score(y_test, y_pred))
    print("F1 Score :", f1_score(y_test, y_pred))
    print("ROC-AUC  :", roc_auc_score(y_test, y_proba))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Evaluate both models
evaluate_model("Logistic Regression", y_test, y_pred_log, y_proba_log)
evaluate_model("Random Forest", y_test, y_pred_rf, y_proba_rf)
```
